```python
import pandas as pd
dib=pd.read_csv("diabetes.csv")
dib
```

```python
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
dib.describe()
```

```python
x=dib.corr()
x
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(x,annot=True)
plt.show()
```

```python
null_values=dib.isnull().sum()
null_values
```

```python
#Replace 0 with No and 1 with Yes
#Although in given dataset this is not required, this is already set 0 and 1
dib=dib.replace({'admitted':{'Yes':1,'No':0}})
dib
```

```python
sns.boxplot(dib['Glucose'])
plt.show()
```

```python
X=dib.iloc[:,:-1]
X
```

```python
Y=dib.iloc[:,-1]
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.3)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
sns.scatterplot(data=dib)
plt.show()
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
x_test
```

```python
y_test
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
score = reg.predict([[8,197,74,0,0,25.9,1.191,39]])
print("Answer is",score)
```

```python
#Actual value and the predicted value
mlr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred})
mlr_diff.head()
```

```python
#Accuracy of Model
from sklearn.metrics import confusion_matrix,accuracy_score
Confusion=confusion_matrix(y_test,y_pred)
Confusion
```

```python
sns.heatmap(Confusion,annot=True,cmap='BuPu')
plt.show()
```

```python
TruePositive=Confusion[1,1]
TrueNegative=Confusion[1,1]
print((TruePositive+TrueNegative)/len(y_test))
```
